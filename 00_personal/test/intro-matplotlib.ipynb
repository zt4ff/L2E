```python
# third method (recommend)
fig, ax = plt.subplots()
ax.plot(x, y); # adds some data
```

## example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# prepare daya
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# save and show
fig.savefig("images/sample-plot.png")
```
