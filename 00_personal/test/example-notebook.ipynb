```python
import pandas as pd

car_sales = pd.read_csv("car-sales.csv")
```

## Manipulating Data

```python
## df.sample - this derives a sample
car_sales.sample(frac=0.2) # 20% sample of the data at random

## sampling picks row at random so doing frac=1 essentially shuffles the dataframe
car_sales_shuffled = car_sales.sample(frac=1)
```
