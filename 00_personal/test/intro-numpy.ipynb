## Dot Product

```python
np.random.seed(0)

mat1 = np.random.randint(10, size=(5, 3))
mat2 = np.random.randint(10, size=(5, 3))

mat3 = np.random.randint(10, size=(2, 3, 4))

print(mat1)
print()
print(mat2)
print()
print(mat3)
```

```python
np.dot(mat1, mat2.T)
```

```python
np.random.seed(0)

sales_amount = np.random.randint(20, size=(5, 3))
sales_amount
```

```python
weekly_sales = pd.DataFrame(sales_amount, index=["Mon", "Tue", "Wed", "Thurs", "Fri"], columns=["Almod butter", "Peanut butter", "Cashew butter"])
 
weekly_sales
```

```python
prices = np.array([10, 8, 12])
butter_prices = pd.DataFrame(prices.reshape(1, 3), index=["Prices"], columns=["Almod butter", "Peanut butter", "Cashew butter"])

butter_prices
```

```python
total_sales = prices.dot(sales_amount.T)
total_sales
```

```python
butter_prices.dot(weekly_sales.T)
```
