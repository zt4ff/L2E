## example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# prepare daya
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# plot data
ax.plot(x, y)

# customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# save and show
fig.savefig("images/sample-plot.png")
```

```python
# make a plot from a dict
nut_butter_prices = {"Almond":10, "Peanut": 8, "Cashew": 12}
fig, ax = plt.subplots()

ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Butter price", xlabel="Products", ylabel="Prices")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
fig, ax = plt.subplots()
x = np.random.randn(1000)

ax.hist(x)
```

```python
np.random.seed(44)
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax1.plot(x, x/0.2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))

plt.show()
```

```python
# subplots option 2
np.random.seed(44)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(x, x/0.2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))

plt.show()
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head(10)
```

```python
heart_disease["age"].plot.hist(bins=10)
```

```python
heart_disease.plot.hist(subplots=True, layout=(7, 2), figsize=(50, 50), bins=50)
plt.show()
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head(10), len(over_50)
```

```python
## plotting directly
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
## oo method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax)
ax.axhline(over_50.chol.mean(), linestyle="--")
```

```python
## oo method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestol")
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50.chol.mean(), linestyle="--");
```

```python
over_50.head()
```

```python
# suplt of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(ncols=1, nrows=2, figsize=(5, 5), sharex=True)

fig.suptitle("Heart Disease Analysis")

scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

ax0.set(ylabel="Cholestol")
ax0.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax0.axhline(over_50.chol.mean(), linestyle="--");


scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

ax1.set(ylabel="Thalach", xlabel="Age")
ax1.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax1.axhline(over_50.thalach.mean(), linestyle="--");

fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap="winter")

ax.set(title="Heart Disease", xlabel="Age", ylabel="Cholestol")
ax.legend(*scatter.legend_elements(), title="Target")

# add horizontal line
ax.axhline(over_50.chol.mean(), linestyle="--");
```
